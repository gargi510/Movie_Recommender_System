# Introduction

The Zee MovieLens dataset contains explicit user–movie rating data for a large catalog of movies, where individual users interact with only a small subset of available items. This leads to a highly sparse user–item interaction matrix, a common challenge in recommender system problems. Such sparsity limits direct inference of user preferences and complicates the task of generating accurate and relevant movie recommendations based solely on historical ratings.


# Problem Statement

Using the Zee MovieLens dataset, the objective is to estimate missing user–movie ratings and generate top-N movie recommendations under sparse data conditions. The problem involves modeling similarity relationships and latent preference structures within the user–movie rating matrix through collaborative filtering techniques and comparing these approaches to evaluate their effectiveness, limitations, and robustness.


# Data Structure and Dictionary

The Zee MovieLens dataset consists of three structured data files containing user ratings, demographic attributes, and movie metadata. These files collectively form the input for exploratory analysis and recommendation modeling.

---

### ratings.dat  
**Format:** `UserID::MovieID::Rating::Timestamp`

| Field | Description |
|------|-------------|
| UserID | Unique user identifier (range: 1–6040) |
| MovieID | Unique movie identifier (range: 1–3952) |
| Rating | Explicit whole-star rating on a 1–5 scale |
| Timestamp | Time of rating recorded in seconds |

Each user has provided a minimum of 20 ratings.

---

### users.dat  
**Format:** `UserID::Gender::Age::Occupation::Zip-code`

| Field | Description |
|------|-------------|
| UserID | Unique user identifier |
| Gender | Gender of the user (M/F) |
| Age | Age group bucket |
| Occupation | Encoded occupation category |
| Zip-code | User zip code |

Demographic information is self-reported and optional.

---

### movies.dat  
**Format:** `MovieID::Title::Genres`

| Field | Description |
|------|-------------|
| MovieID | Unique movie identifier |
| Title | Movie title including release year |
| Genres | Pipe-separated list of movie genres |

Genres include standard categories such as Action, Comedy, Drama, Romance, Sci-Fi, Thriller, and others.


# Approach

The project follows a structured approach combining exploratory analysis and multiple collaborative filtering techniques to build and evaluate a movie recommender system.

### Exploratory Data Analysis (EDA)
Exploratory analysis at the **user, movie, and rating levels** includes analyzing the number of ratings per user, movie popularity distribution, and overall rating behavior. The EDA identifies sparsity patterns and informs decisions related to similarity measures, imputation strategy, and model selection.

### Similarity-Based Collaborative Filtering
Item-based collaborative filtering recommends movies similar to a given movie based on rating patterns. Pearson correlation captures linear similarity between movies with overlapping ratings. User-mean imputation addresses missing values, and cosine similarity is computed on the imputed user–movie matrix. A k-Nearest Neighbors (kNN) approach retrieves the top-N most similar movies.

### Matrix Factorization
A collaborative matrix factorization model using Alternating Least Squares (ALS) decomposes the user–movie rating data into low-dimensional latent representations of users and movies. Evaluation uses RMSE and MAPE on a held-out test set. Learned embeddings are further analyzed by computing user–user and item–item similarities and visualizing item embeddings using PCA.

### Neural Network–Based Recommendation
A feedforward autoencoder trains on the user–movie rating matrix to capture non-linear patterns in user preferences. The encoder provides latent user embeddings, and cosine similarity over learned representations generates movie recommendations.

Comparison of approaches assesses effectiveness under sparse data conditions.

The descriptive analysis of the ratings dataset shows that there are 6040 unique users and 3706 unique movies. Ratings are on a 1–5 scale, with 4 being the most frequent rating. The `Timestamp` column has 458,455 unique values, indicating multiple ratings recorded at the same time, and some users and movies appear more frequently, reflecting rating activity skewness.

```python
df_ratings['Rating'] = pd.to_numeric(df_ratings['Rating'], errors='coerce')
df_ratings['Timestamp'] = pd.to_numeric(df_ratings['Timestamp'], errors='coerce')
```

The `Rating` and `Timestamp` columns have been converted from object to numeric types to enable statistical analysis and exploratory data analysis.


```python
df_ratings_eda = df_ratings.copy()

df_ratings_eda['Timestamp'] = pd.to_datetime(df_ratings_eda['Timestamp'], unit='s')
df_ratings_eda['year'] = df_ratings_eda['Timestamp'].dt.year
df_ratings_eda['month'] = df_ratings_eda['Timestamp'].dt.month
df_ratings_eda['dayofweek'] = df_ratings_eda['Timestamp'].dt.dayofweek
df_ratings_eda['hour'] = df_ratings_eda['Timestamp'].dt.hour

user_stats = df_ratings_eda.groupby('UserID')['Rating'].agg(
    user_mean_rating='mean',
    user_rating_count='count'
).reset_index()
df_ratings_eda = df_ratings_eda.merge(user_stats, on='UserID', how='left')

movie_stats = df_ratings_eda.groupby('MovieID')['Rating'].agg(
    movie_mean_rating='mean',
    movie_rating_count='count'
).reset_index()
df_ratings_eda = df_ratings_eda.merge(movie_stats, on='MovieID', how='left')

df_ratings_eda.head()
```

A copy of the ratings DataFrame is created (`df_ratings_eda`) to preserve the original data for model training.

The `Timestamp` column is converted to datetime format, from which new temporal features are extracted: `year`, `month`, `dayofweek`and `hour`

Additionally, user-level statistics (`user_mean_rating`, `user_rating_count`) and movie-level statistics (`movie_mean_rating`, `movie_rating_count`) are computed and merged back into the DataFrame.

These engineered features facilitate detailed exploratory analysis of rating patterns by time, user activity, and movie popularity.

**Features created**

- **year**: Year when the rating was recorded  
- **month**: Month of the rating  
- **dayofweek**: Day of the week of the rating (0=Monday, 6=Sunday)  
- **hour**: Hour of the day when the rating was made  
- **user_mean_rating**: Average rating given by the user  
- **user_rating_count**: Total number of ratings made by the user  
- **movie_mean_rating**: Average rating received by the movie  
- **movie_rating_count**: Total number of ratings received by the movie

## Genre Preference Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(24, 6))

genre_cols = [c for c in df_movies.columns[3:] if c != 'genre_count']

df_ratings_movies = df_ratings.merge(
    df_movies[['Movie ID'] + list(genre_cols)],
    left_on='MovieID', right_on='Movie ID', how='left'
).drop(columns=['Movie ID'])

df_gender = df_ratings_movies.merge(df_users[['UserID','Gender']], on='UserID', how='left')

avg_rating_genre = pd.Series({
    g: df_ratings_movies[df_ratings_movies[g]==1]['Rating'].mean()
    for g in genre_cols
}).sort_values(ascending=False)

count_rating_genre = pd.Series({
    g: df_ratings_movies[df_ratings_movies[g]==1]['Rating'].count()
    for g in genre_cols
}).sort_values(ascending=False)

gender_genre = df_gender.melt(
    id_vars=['UserID','Gender','Rating'],
    value_vars=genre_cols,
    var_name='Genre', value_name='is_genre'
)
gender_genre = gender_genre[gender_genre['is_genre']==1]
gender_genre = gender_genre.groupby(['Genre','Gender'])['Rating'].mean().unstack()
gender_genre = gender_genre.astype(float)

avg_rating_genre.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Average Rating per Genre')
axes[0].set_ylabel('Average Rating')

count_rating_genre.plot(kind='bar', ax=axes[1], color='lightgreen')
axes[1].set_title('Number of Ratings per Genre')
axes[1].set_ylabel('Count')

gender_genre.plot(kind='bar', ax=axes[2], color=['lightpink','dodgerblue'])
axes[2].set_title('Gender-wise Avg Rating per Genre')
axes[2].set_ylabel('Average Rating')
axes[2].legend(title='Gender', labels=['Female','Male'])

plt.tight_layout()
plt.show()
```

This analysis links user behavior with movie content to inform recommendation strategies:  

- **Average rating per genre:** Highest in Film-Noir, Documentary, and War; lowest in Horror. Popular genres like Comedy, Drama, and Action have moderate ratings, showing the **"Quality vs. Popularity" paradox**.  
- **Genre popularity vs. quality:** Niche genres receive fewer ratings but higher scores; mass-appeal genres are widely rated with average scores.  
- **Gender alignment:** Men and women have similar rating patterns; small differences appear in Musicals, Romance, Film-Noir, and Westerns. Horror is low-rated across both genders.  
- **Rating volume:** Males provide ~3x more ratings than females; female ratings are slightly higher on average.  

Insights help bridge content and user behavior for content-aware or hybrid recommendation approaches.

# Model Building and Evaluation

## Pearson Correlation

```python
df_merged = df_ratings.merge(
    df_movies[['Movie ID', 'Title']],
    left_on='MovieID',
    right_on='Movie ID',
    how='inner'
)

df_merged = df_merged[['UserID', 'Title', 'Rating']]
df_merged.head()
```

To compute movie–movie similarity using Pearson correlation, the ratings data is merged with the movie metadata to associate each rating with its corresponding movie title. This creates a user–movie–rating table (`UserID`, `Title`, `Rating`), which is required to construct the user–movie matrix and measure similarity between movies based on shared user rating patterns for generating top-5 similar movie recommendations.


```python
user_movie_matrix = df_merged.pivot_table(
    index='UserID',
    columns='Title',
    values='Rating'
)
user_movie_matrix.head()
```

A user–movie interaction matrix is created using a pivot table, with users as rows, movies as columns, and ratings as values. This matrix serves as the foundational structure for similarity-based collaborative filtering and subsequent recommendation modeling.

We leave NaNs before computing Pearson correlation because they represent missing ratings, not actual user preferences. Pearson correlation naturally ignores these missing values and uses only users who rated both movies, which preserves true behavioral similarity. Imputing values would introduce artificial ratings and bias the similarity scores.

```python
def recommend_using_pearson(user_movie_matrix, movie_name, top_n=5):

    movie_similarity = user_movie_matrix.corr(method='pearson')

    if movie_name not in movie_similarity.columns:
        return f"Movie '{movie_name}' not found."

    similarity_scores = movie_similarity[movie_name].drop(movie_name).dropna()

    top_movies = similarity_scores.sort_values(ascending=False).head(top_n)

    return top_movies
```

A function is defined to compute movie–movie similarity using Pearson correlation. Given a movie name as input, the function calculates correlation scores between the selected movie and all other movies based on user rating patterns, filters out invalid or missing values, and returns the top-5 most similar movies with the highest correlation scores.


## Cosine Similarity

```python
def recommend_using_cosine_similarity(user_movie_matrix, movie_name, top_n=5, item_embeddings=None):
    if item_embeddings is not None:
        item_sim_matrix = cosine_similarity(item_embeddings)
        item_similarity = pd.DataFrame(item_sim_matrix, index=user_movie_matrix.columns, columns=user_movie_matrix.columns)
        user_sim_matrix = None
    else:
        user_movie_csr = csr_matrix(user_movie_matrix.values)
        item_sim_matrix = cosine_similarity(user_movie_csr.T)
        item_similarity = pd.DataFrame(item_sim_matrix, index=user_movie_matrix.columns, columns=user_movie_matrix.columns)
        user_sim_matrix = cosine_similarity(user_movie_csr)
        user_similarity = pd.DataFrame(user_sim_matrix, index=user_movie_matrix.index, columns=user_movie_matrix.index)
        print("Item-Item Cosine Similarity Matrix (top 5 columns):")
        print(item_similarity.iloc[:, :5].head())
        print("\nUser-User Cosine Similarity Matrix (top 5 users):")
        print(user_similarity.iloc[:, :5].head())

    if movie_name not in item_similarity.columns:
        return f"Movie '{movie_name}' not found."

    similarity_scores = item_similarity[movie_name].drop(movie_name).sort_values(ascending=False)
    top_movies = similarity_scores.head(top_n)

    return top_movies
```

A function is implemented to recommend movies using cosine similarity. It computes item–item similarity either directly from the user–movie rating matrix or from precomputed item embeddings, if provided. The function then identifies the movies most similar to the given movie based on cosine similarity scores and returns the top-5 recommendations, handling cases where the requested movie is not present in the dataset.


## KNN with Cosine Similarity

```python
def recommend_using_knn_cosine(user_movie_matrix, movie_name, top_n=5):

    movie_vectors = user_movie_matrix.T
    movie_vectors_csr = csr_matrix(movie_vectors.values)

    knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
    knn_model.fit(movie_vectors_csr)

    if movie_name not in movie_vectors.index:
        return f"Movie '{movie_name}' not found."

    movie_idx = movie_vectors.index.get_loc(movie_name)
    distances, indices = knn_model.kneighbors(
        movie_vectors.iloc[movie_idx, :].values.reshape(1, -1),
        n_neighbors=top_n + 1
    )

    similar_movies = []
    similarity_scores = []

    for i, dist in zip(indices[0], distances[0]):
        if i != movie_idx:
            similar_movies.append(movie_vectors.index[i])
            similarity_scores.append(1 - dist)

    return pd.Series(similarity_scores, index=similar_movies)
```

A KNN-based recommendation function is defined using cosine distance. The user–movie matrix is transposed so that each movie is represented as a vector of user ratings, then converted to a sparse matrix for efficiency. A Nearest Neighbors model with cosine similarity is trained on these movie vectors. Given a movie name, the function finds its nearest neighbors in the rating space and returns the top-5 most similar movies, along with their similarity scores, while excluding the movie itself and handling cases where the movie is not found.


## Recommendation

```python
movie_name = "Liar Liar"

print("Top 5 similar movies (Pearson):")
print(recommend_using_pearson(user_movie_matrix, movie_name))

#imputation
user_movie_matrix = user_movie_matrix.apply(lambda row: row.fillna(row.mean()), axis=1)

print("\nTop 5 similar movies (Cosine, user-mean imputation, CSR):")
print(recommend_using_cosine_similarity(user_movie_matrix, movie_name))

print("\nTop 5 similar movies (KNN + Cosine):")
print(recommend_using_knn_cosine(user_movie_matrix, movie_name))
```

The results highlight the limitations of similarity-based collaborative filtering on a highly sparse matrix (~95% sparsity).  

**Pearson correlation** is unreliable, as perfect similarity scores (1.0) indicate very few overlapping ratings, producing inflated and noisy similarities.  

**Cosine similarity** after user-mean imputation appears stable but yields uniformly high scores (~0.997), showing loss of discrimination due to reduced variance.  

**KNN with cosine** produces results identical to item–item cosine, confirming it cannot overcome sparsity.  

The **user–user cosine similarity** matrix shows extremely high similarity values across most users, indicating that mean imputation makes user profiles overly similar and reduces personalization.  

The **item–item cosine similarity** matrix also exhibits near-uniformly high scores, suggesting weak signal differentiation between movies.  

Overall, both matrices confirm that sparsity and imputation flatten the interaction space, limiting the effectiveness of neighborhood-based collaborative filtering and justifying the move to Matrix Factorization and Neural Network models.


## Matrix Factorization Approach

```python
df_mf = df_ratings[['UserID', 'MovieID', 'Rating']].copy()
df_mf['UserID'] = df_mf['UserID'].astype(int)
df_mf['MovieID'] = df_mf['MovieID'].astype(int)
df_mf['Rating'] = df_mf['Rating'].astype(float)

user_ids = df_mf['UserID'].unique()
movie_ids = df_mf['MovieID'].unique()
user_map = {u: i for i, u in enumerate(user_ids)}
movie_map = {m: i for i, m in enumerate(movie_ids)}
inv_user_map = {v: k for k, v in user_map.items()}
inv_movie_map = {v: k for k, v in movie_map.items()}

df_mf['UserIdx'] = df_mf['UserID'].map(user_map)
df_mf['ItemIdx'] = df_mf['MovieID'].map(movie_map)
rm_raw = df_mf[['UserIdx', 'ItemIdx', 'Rating']].copy()
rm_raw.columns = ['UserId', 'ItemId', 'Rating']

train_df, test_df = train_test_split(rm_raw, test_size=0.2, random_state=42)

d = 4
model = CMF(method="als", k=d, lambda_=0.1, user_bias=False, item_bias=False, verbose=True)
model.fit(train_df)

movie_titles = dict(zip(df_movies['Movie ID'].astype(int), df_movies['Title']))


def recommend_cmf(user_id, n=5):
    if user_id not in user_map:
        return []
    u_idx = user_map[user_id]
    top_items_idx = model.topN(user=u_idx, n=n)
    return [movie_titles[inv_movie_map[i]] for i in top_items_idx]


print("Top 5 recommendations for user 1:")
print(recommend_cmf(user_id=1, n=5))

test_preds = []
test_actuals = []

for idx, row in test_df.iterrows():
    try:
        pred = model.predict(user=int(row['UserId']), item=int(row['ItemId']))
        test_preds.append(pred)
        test_actuals.append(row['Rating'])
    except:
        continue

test_preds = np.array(test_preds)
test_actuals = np.array(test_actuals)

rmse = np.sqrt(mean_squared_error(test_actuals, test_preds))
mape = mean_absolute_percentage_error(test_actuals, test_preds)

print(f"\nTest RMSE: {rmse:.4f}")
print(f"Test MAPE: {mape:.4f}")

item_sim = cosine_similarity(model.B_)
user_sim = cosine_similarity(model.A_)

n_items = model.B_.shape[0]
n_users = model.A_.shape[0]

movie_id_list = [int(inv_movie_map[i]) for i in range(n_items)]
user_id_list = [int(inv_user_map[i]) for i in range(n_users)]

item_sim_df = pd.DataFrame(item_sim, index=movie_id_list, columns=movie_id_list)
user_sim_df = pd.DataFrame(user_sim, index=user_id_list, columns=user_id_list)


def top_similar_items(movie_id, n=5):
    movie_id = int(movie_id)
    if movie_id not in item_sim_df.index:
        print(f"Movie ID {movie_id} not found in similarity matrix")
        return []
    sim_scores = item_sim_df.loc[movie_id].drop(movie_id).sort_values(ascending=False)
    top_items = sim_scores.head(n).index
    return [movie_titles.get(int(mid), f"Movie {mid}") for mid in top_items]


def top_similar_users(user_id, n=5):
    user_id = int(user_id)
    if user_id not in user_sim_df.index:
        return []
    sim_scores = user_sim_df.loc[user_id].drop(user_id).sort_values(ascending=False)
    return list(sim_scores.head(n).index)

movie_id = int(df_movies[df_movies['Title'] == movie_name]['Movie ID'].values[0])
print(f"\nTop 5 movies similar to '{movie_name}':")
similar_movies = top_similar_items(movie_id, n=5)
for movie in similar_movies:
    print(f"  - {movie}")

# PCA visualization
pca = PCA(n_components=2)
B_2D = pca.fit_transform(model.B_)
plt.figure(figsize=(10, 6))
plt.scatter(B_2D[:, 0], B_2D[:, 1], alpha=0.5)
plt.title("2D PCA of Item Embeddings (d=4 MF)")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

Top 5 movies similar to 'Liar Liar':
  - Man Who Knew Too Much, The
  - Midaq Alley (Callejón de los milagros, El)
  - Blues Brothers 2000
  - Better Living Through Circuitry
  - Monty Python's Life of Brian

- **ALS-based Matrix Factorization** successfully captured latent patterns from the sparse user-item matrix.
- **Top-N recommendations** align with user preferences; test RMSE (0.8775) and MAPE (0.2626) indicate good predictive performance.
- **Movie similarity** (e.g., *Liar Liar*) is derived from latent features, reflecting deeper relationships beyond co-ratings.
- **4D latent space reduced to 2D** shows a dense central cluster with slight outliers, indicating most items share overlapping characteristics.
- No distinct clusters appear, suggesting continuous item similarities rather than strictly categorical groupings.
- **Matrix Factorization overcomes sparsity** limitations of neighborhood methods, providing robust recommendations and meaningful similarity insights.


## Neural Network Approach

```python
R_full = user_movie_matrix.values.astype(np.float32)
mask_full = R_full != 0

idx = np.random.choice(R_full.shape[0], 5000, replace=False)
R, mask = R_full[idx], mask_full[idx]

inp = layers.Input(shape=(R.shape[1],))
x = layers.Dense(32, activation="relu")(inp)
z = layers.Dense(4, activation="linear")(x)
x = layers.Dense(32, activation="relu")(z)
out = layers.Dense(R.shape[1], activation="linear")(x)

model = models.Model(inp, out)
model.compile(optimizer=optimizers.Adam(0.001), loss="mse")
model.fit(R, R, epochs=50, batch_size=64, verbose=1)

R_pred = model.predict(R, verbose=0)
mape = np.mean(np.abs((R[mask] - R_pred[mask]) / R[mask]))
print(f"MAPE: {mape:.4f}")

decoder_w = model.layers[-1].get_weights()[0]
proj_w = model.layers[-2].get_weights()[0]
item_embeddings = (proj_w @ decoder_w).T

item_embeddings_df = pd.DataFrame(
    item_embeddings,
    index=user_movie_matrix.columns
)

top_movies = recommend_using_cosine_similarity(
    user_movie_matrix,
    movie_name,
    top_n=5,
    item_embeddings=item_embeddings_df.values
)

print(top_movies)
```

- **Neural Network Autoencoder** was trained to learn low-dimensional embeddings for movies.  
- **Best performance** achieved with `embedding_dim=4`, `hidden_size=32`, `lr=0.001`, `batch_size=64` resulting in **MAPE = 0.02** on the training subset.  
- The autoencoder compresses the user-item matrix into a **latent representation** and reconstructs ratings, capturing complex, non-linear relationships between users and items.  
- The learned **item embeddings** were passed to the cosine similarity function to retrieve top-5 similar movies for a given title.  
- Example for *Liar Liar*:  
  - *West Side Story* – 0.984  
  - *Searching for Bobby Fischer* – 0.976  
  - *Blood and Sand (Sangre y Arena)* – 0.970  
  - *Stealing Beauty* – 0.970  
  - *Foxfire* – 0.965  
- **Top-5 similarity using embeddings** achieved **MAPE = 0.02**, showing that the autoencoder effectively captures nuanced movie relationships.  
- This method **outperforms traditional neighborhood methods** and is robust to sparsity, producing high-quality, content-aware recommendations.

  - Pearson correlation produces perfect scores (1.0) due to minimal overlap.

  - KNN-based cosine similarity mirrors item-item results.

  - Successfully modeled latent factors with RMSE = 0.8775 and MAPE = 0.2626.

- **Recommendations:**  
  1. **Use latent factor models** (Matrix Factorization or Autoencoders) rather than neighborhood-based CF due to high sparsity.  
  2. **Incorporate content features** (genre, release year) to improve recommendations for niche or low-rated movies.  
  3. **Leverage time-aware modeling** to account for recency, seasonal, and weekly trends in user engagement.

  5. **Deploy Autoencoder embeddings** for similarity-based recommendations to achieve high precision for top-N items.  
  6. **Cold-start consideration:** While cold-start is not a major issue here due to each user having ≥20 ratings, any new user or movie would require content-based features or hybrid approaches to generate reliable recommendations.  
  7. Monitor trends for **popular vs. niche genres** to balance recommendation diversity and quality perception.

- **Conclusion:**  
  The combination of MF and Neural Network Autoencoder provides a robust framework for generating personalized, high-quality recommendations in sparse, large-scale datasets, outperforming traditional CF approaches and capturing complex user-item relationships, while the minimal cold-start risk allows focus on latent factor modeling.
